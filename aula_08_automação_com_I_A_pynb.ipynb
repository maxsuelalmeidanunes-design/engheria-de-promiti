Missão 01: automação de arquivos

```python
import os # Importa o módulo 'os' para interagir com o sistema operacional, como listar arquivos e criar pastas.
import shutil # Importa o módulo 'shutil' para operações de alto nível em arquivos e coleções de arquivos, como mover.

# --- Configurações ---

# ATENÇÃO: Substitua 'caminho/para/seus/arquivos/baguncados' pelo caminho real da pasta onde estão seus arquivos.
source_directory = 'test_files_source' # Agora aponta para a pasta de arquivos de teste criada automaticamente.

# ATENÇÃO: Substitua 'caminho/para/o/diretorio/organizado' pelo caminho real onde você deseja criar as novas pastas organizadas.
destination_directory = 'test_files_organized' # Agora aponta para a pasta de destino para os arquivos organizados.

# --- Lógica de Organização ---

# 1. Verifica se o diretório de destino existe. Se não existir, ele é criado.
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)
    print(f"Diretório de destino '{destination_directory}' criado.")

# 2. Itera sobre todos os itens (arquivos e pastas) no diretório de origem.
for filename in os.listdir(source_directory):
    # Constrói o caminho completo para o item atual no diretório de origem.
    file_path = os.path.join(source_directory, filename)

    # 3. Verifica se o item atual é de fato um arquivo (e não uma subpasta).
    if os.path.isfile(file_path):
        # 4. Obtém a extensão do arquivo (ex: '.pdf', '.jpg').
        # os.path.splitext retorna uma tupla (nome_base, extensão), pegamos o segundo elemento [1].
        # .lower() é usado para padronizar as extensões (ex: '.JPG' vira '.jpg').
        file_extension = os.path.splitext(filename)[1].lower()

        # 5. Define o nome da pasta de destino com base na extensão.
        if not file_extension: # Se o arquivo não tem extensão (ex: 'meuarquivo_sem_extensao').
            target_folder_name = 'Sem Extensão'
        else: # Para arquivos com extensão (ex: '.pdf', '.docx').
            # Remove o ponto inicial da extensão para usar como nome da pasta (ex: '.pdf' vira 'pdf').
            target_folder_name = file_extension[1:]

        # 6. Constrói o caminho completo para a pasta de destino específica para esta extensão.
        target_folder_path = os.path.join(destination_directory, target_folder_name)

        # 7. Verifica se a pasta para esta extensão já existe no diretório de destino. Se não existir, ela é criada.
        if not os.path.exists(target_folder_path):
            os.makedirs(target_folder_path)
            print(f"Pasta '{target_folder_name}' criada em '{destination_directory}'.")

        # 8. Constrói o caminho completo para onde o arquivo será movido.
        destination_file_path = os.path.join(target_folder_path, filename)

        # 9. Adiciona tratamento para arquivos já existentes: renomeia com sufixo se o nome já existir.
        base_name, ext = os.path.splitext(filename)
        counter = 1
        original_filename = filename # Guarda o nome original para a mensagem
        while os.path.exists(destination_file_path):
            # Se o arquivo já existe, tenta um novo nome com um contador (ex: 'arquivo_copy1.ext')
            filename = f"{base_name}_copy{counter}{ext}"
            destination_file_path = os.path.join(target_folder_path, filename)
            counter += 1

        # 10. Move o arquivo do diretório de origem para o pasta de destino correspondente.
        try:
            shutil.move(file_path, destination_file_path)
            if original_filename != filename:
                print(f"Arquivo '{original_filename}' já existia. Movido como '{filename}' para '{target_folder_name}'.")
            else:
                print(f"Arquivo '{filename}' movido para '{target_folder_name}'.")
        except Exception as e:
            print(f"Erro ao mover o arquivo '{original_filename}': {e}")

print("\nOrganização de arquivos concluída!")
print("Lembre-se de alterar as variáveis 'source_directory' e 'destination_directory' com os caminhos corretos antes de executar.")
```

```python
import os
import shutil

source_directory = 'test_files_source'
destination_directory = 'test_files_organized'

# Remove a pasta de origem e destino se existirem para garantir um teste limpo
if os.path.exists(source_directory):
    shutil.rmtree(source_directory)
    print(f"Diretório '{source_directory}' existente removido.")

if os.path.exists(destination_directory):
    shutil.rmtree(destination_directory)
    print(f"Diretório '{destination_directory}' existente removido.")

# Cria a pasta de origem se ela não existir
os.makedirs(source_directory, exist_ok=True)
print(f"Diretório '{source_directory}' criado para arquivos de teste.")

# Lista de arquivos de teste para criar
files_to_create = [
    'documento.pdf',
    'imagem.jpg',
    'relatorio.docx',
    'planilha.xlsx',
    'script.py',
    'foto.png',
    'apresentacao.pptx',
    'audio.mp3',
    'video.mp4',
    'texto_sem_extensao',
    'outro_documento.pdf',
    'copia_imagem.jpg' # Arquivo duplicado para testar renomeação
]

# Cria os arquivos de teste dentro da pasta de origem
for filename in files_to_create:
    file_path = os.path.join(source_directory, filename)
    with open(file_path, 'w') as f:
        f.write(f'Conteúdo de teste para {filename}')
    print(f"Arquivo '{filename}' criado em '{source_directory}'.")

print("\nArquivos de teste criados com sucesso na pasta 'test_files_source'!")
```

Missão 2: consulta simples a APIs

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API pública VIACEP para obter informações de endereço.
    Args:
        cep (str): O número do CEP a ser consultado (pode conter não-dígitos).
    Returns:
        dict or None: Um dicionário com os dados do endereço formatados, ou None se ocorrer um erro.
    """
    # Remove qualquer caractere não numérico do CEP para garantir 8 dígitos numéricos
    cep = ''.join(filter(str.isdigit, cep))

    if len(cep) != 8:
        print("Erro: O CEP deve conter exatamente 8 dígitos numéricos.")
        return None

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Faz a requisição HTTP para a API
        response = requests.get(url)
        response.raise_for_status()  # Lança uma exceção para códigos de status HTTP de erro (4xx ou 5xx)

        data = response.json()

        # Verifica se a API retornou um erro (ex: CEP inválido)
        if 'erro' in data and data['erro']:
            print(f"Erro: CEP '{cep}' não encontrado ou inválido.")
            return None

        # Formata as informações de endereço em um dicionário
        endereco_formatado = {
            "CEP": data.get('cep'),
            "Logradouro": data.get('logradouro'),
            "Complemento": data.get('complemento'),
            "Bairro": data.get('bairro'),
            "Cidade": data.get('localidade'),
            "Estado": data.get('uf'),
            "IBGE": data.get('ibge'),
            "GIA": data.get('gia'),
            "DDD": data.get('ddd'),
            "SIAFI": data.get('siafi')
        }
        return endereco_formatado

    except requests.exceptions.HTTPError as http_err:
        print(f"Erro HTTP ao consultar o CEP {cep}: {http_err}")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Erro de conexão ao consultar o CEP {cep}: {conn_err}. Verifique sua conexão com a internet.")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Erro de tempo limite ao consultar o CEP {cep}: {timeout_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"Ocorreu um erro inesperado na requisição para o CEP {cep}: {req_err}")
    except ValueError as val_err:
        print(f"Erro ao processar a resposta JSON para o CEP {cep}: {val_err}")
    except Exception as e:
        print(f"Ocorreu um erro desconhecido ao consultar o CEP {cep}: {e}")

    return None

# --- Exemplo de uso interativo ---

# Solicita ao usuário que insira um CEP
cep_digitado = input("Digite o CEP (apenas números): ")

# Consulta o CEP
endereco = consultar_cep(cep_digitado)

# Exibe o resultado de forma legível
if endereco:
    print("\n--- Informações do Endereço ---")
    for chave, valor in endereco.items():
        if valor: # Exibe apenas campos com valor preenchido
            print(f"{chave}: {valor}")
else:
    print("Não foi possível obter informações para o CEP fornecido.")
```
