### Calculadora Simples

Esta calculadora permite somar ou subtrair dois números.

```python
# Bloco principal para a calculadora
try:
    # 1. Solicita os dois números ao usuário
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
# 2. Lida com entradas inválidas (não numéricas)
except ValueError:
    print("Entrada inválida. Por favor, digite números válidos.")
else:
    # 3. Solicita a operação ao usuário e converte para minúsculas
    operacao_input = input("Qual operação você deseja realizar? (somar/subtrair): ").lower()

    # Inicializa variáveis para o resultado e o nome da operação
    resultado = None
    operacao_nome = ""

    # 4. Verifica a operação desejada e calcula o resultado
    if operacao_input == "somar":
        resultado = num1 + num2
        operacao_nome = "soma"
    elif operacao_input == "subtrair":
        resultado = num1 - num2
        operacao_nome = "subtração"

    # 5. Exibe o resultado ou uma mensagem de operação inválida
    if resultado is not None:
        print(f"O resultado da {operacao_nome} é: {resultado}")
    else:
        print("Operação inválida. Por favor, escolha 'somar' ou 'subtrair'.")
```
